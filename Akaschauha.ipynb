```python
import string

def encryption(plain, shift):
    alphabet = string.ascii_lowercase
    cipher = ""
    for char in plain:
        if char.isalpha():
            if char.islower():
                position = alphabet.index(char)
                new = (position + shift) % 26
                cipher += alphabet[new]
            else:
                position = alphabet.index(char.lower())
                new = (position + shift) % 26
                cipher += alphabet[new].upper()
        else:
            cipher += char
    print(f"The text after encryption is: {cipher}")

def decryption(cipher, shift):
    alphabet = string.ascii_lowercase
    plain = ""
    for char in cipher:
        if char.isalpha():
            if char.islower():
                position = alphabet.index(char)
                new = (position - shift) % 26
                plain += alphabet[new]
            else:
                position = alphabet.index(char.lower())
                new = (position - shift) % 26
                plain += alphabet[new].upper()
        else:
            plain += char
    print(f"The text after decryption is: {plain}")

end = False
while not end:
    what = input("Type 'encrypt' for encryption and 'decrypt' for decryption: ")
    text = input("Type your message: ")
    shift = int(input("Enter shift: "))
    if what == "encrypt":
        encryption(plain=text, shift=shift)
    elif what == "decrypt":
        decryption(cipher=text, shift=shift)
    play = input('Type "yes  to continue "or "no for discontinue": ')
    if play.lower() == 'no':
        end = True
        print("Have a nice day!")
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class ToDoList:
    def __init__(self):
        self.head = None

    def add_task(self, task):
        new_node = Node(task)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def view_tasks(self):
        if not self.head:
            print("No tasks in the to-do list.")
        else:
            print("Tasks in To-Do List:")
            current = self.head
            index = 1
            while current:
                print(f"{index}. {current.data}")
                current = current.next
                index += 1

    def remove_task(self, index):
        try:
            index = int(index)
            if index == 1:
                self.head = self.head.next
            else:
                prev = None
                current = self.head
                count = 1
                while current and count != index:
                    prev = current
                    current = current.next
                    count += 1
                if current:
                    prev.next = current.next
                else:
                    print("Invalid task index.")
        except ValueError:
            print("Invalid input. Please enter a valid index.")

    def save_todo_list(self, filename):
        with open(filename, 'w') as f:
            current = self.head
            while current:
                f.write(current.data + '\n')
                current = current.next
        print("To-do list saved successfully!")

    def load_todo_list(self, filename):
        try:
            with open(filename, 'r') as f:
                lines = f.readlines()
                for line in lines:
                    self.add_task(line.strip())
            print("To-do list loaded successfully!")
        except FileNotFoundError:
            print("File not found. Creating a new to-do list.")

def main():
    todo_list = ToDoList()
    filename = "todo_list.txt"
    todo_list.load_todo_list(filename)

    while True:
        print("\n===== To-Do List Manager =====")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Remove Task")
        print("4. Save To-Do List")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            task = input("Enter the task: ")
            todo_list.add_task(task)
        elif choice == '2':
            todo_list.view_tasks()
        elif choice == '3':
            index = input("Enter the index of the task to remove: ")
            todo_list.remove_task(index)
        elif choice == '4':
            todo_list.save_todo_list(filename)
        elif choice == '5':
            todo_list.save_todo_list(filename)
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```

```python
#use classes for the final project:Wq

# A dict to store users data
import json


users = {}

# Funct to register a new user
def registeration(username, password):

  if username in users:
    print("Username already taken")

  else:    #Making a dictionary inside users for each unique user to store thier pass and posts
    users[username] = {"password": password, "posts": []}
    print(f"User '{username}' created successfully!")   #usinf f strings here to give value to username at runtime

# Function to login a user
def login(username, password):

  if username not in users:
    print("Invalid username!")
    return None

  # checking for correct password
  elif users[username]["password"] != password:
    print("Invalid password!")
    return None

  else:
    print(f"Welcome back, {username}!")  # Again f strings !!!
    return username

def save_user_content(username, filename):
    if username not in users:
        print("Invalid username!")
        return
    with open(filename, 'w') as f:
        json.dump(users[username]["posts"], f)
    print(f"User '{username}' content saved successfully!")


# Function to load user-generated content from a file
def load_user_content(username, filename):
    if username not in users:
        print("Invalid username!")
        return
    try:
        with open(filename, 'r') as f:
            users[username]["posts"] = json.load(f)
        print(f"User '{username}' content loaded successfully!")
    except FileNotFoundError:
        print("File not found. No user content loaded.")



# Function to create new blog post
def create_new_post(username, title, content):
  if username not in users:
    print("Please login to create a post!")
  else:
    users[username]["posts"].append({"title": title, "content": content})   #appending the post list with title and content whoch is accesed through users dictionary
    print(f"Post '{title}' created successfully!")

# Function to view all posts by an individual user
def view(username):
  if username not in users:
    print("Invalid username!")

  elif not users[username]["posts"]:        # check whether users have any posts stored , compares users[username] and ["posts"]
    print(f"{username} has no posts yet!")

  else:
    print(f"Posts by {username}:")
    for post in users[username]["posts"]:
      print(f"\tTitle: {post['title']}")
      print(f"\tContent: {post['content']}")
      print("---")

# Function to update a blog post
def update_posts(username, post_title, new_content):

  if username not in users:
    print("Invalid username!")
  else:
    found = 0
    for post in users[username]["posts"]:
      if post["title"] == post_title:  # to know update which blogpost
        post["content"] = new_content
        print(f"Post '{post_title}' updated successfully!")
        found = 1
        break
    if not found:
      print(f"Post '{post_title}' not found!")

# Function to delete a blog post
def delete_post_(username, post_title):

    if username not in users:
        print("Invalid username!")

    else:
        found = 0
        posts = users[username]["posts"]
        for i in range(len(posts)):
            if posts[i]["title"] == post_title:  #which post to delete identified by post title
                del posts[i]   #deleting posts line by line
                print(f"Post '{post_title}' deleted successfully!")
                found = 1
                break
        if not found:
            print(f"Post '{post_title}' not found!")


def delete_user(username):

  if username not in users:
    return f"User '{username}' not found!"

  else:
    del users[username]
    return f"User '{username}' deleted successfully!"


def logout(username):
  print(f"Goodbye Hope to see you again, {username}!")
  loggedin_user = None
  from cms import  registeration
from cms import login
from cms import create_new_post
from cms import view
from cms import update_posts
from cms import delete_post_
from cms import logout
from cms import delete_user
from cms import save_user_content




def main():

  filename = "cms.json"

  loggedin_user  = None  #checks for already logged in user since main run everytimes


  while True:
    if not loggedin_user:
      print("1. Register")
      print("2. Login")
      print("3. Delete user")
      print("4. Exit cms")
      choice = input("Enter your choice: ")
      if choice == "1":
        username = input("Username: ")
        password = input("Password: ")
        registeration(username, password)
      elif choice == "2":
        username = input("Username: ")
        password = input("Password: ")
        loggedin_user = login(username, password)
      elif choice == "3":
        username = input("User to be deleted")
        delete_user(username)
        return("User deleted succesfully")

      elif choice == "4":
        print("Exiting the CMS.")
        break
    else:
      print("1. Create Post")
      print("2. View Posts")
      print("3. Update Post")
      print("4. Delete Post")
      print("5. Save Post")
      print("6. Logout")
      print("7. Delete")

      choice = input("Enter your choice: ")

      if choice == "1":
        title = input("Post title: ")
        content = input("Post content: ")
        create_new_post(loggedin_user, title, content)

      elif choice == "2":
        view(loggedin_user)

      elif choice == "3":
        post_title = input("Enter post title to update: ")
        new_content = input("Enter new content: ")
        update_posts(loggedin_user, post_title, new_content)

      elif choice == "4":
        post_title = input("Enter post title to delete: ")
        delete_post_(loggedin_user, post_title)

      elif choice == '5':
            if username:
              save_user_content(username, filename)
            else:
                print("Please login to save the to-do list!")

      elif choice == "6":
        logout(loggedin_user)
        loggedin_user = None               #loop breaks


      elif choice == "7":
        username = input("Enter user to be deleted")
        delete_user(username)

      else:
        print("Invalid choice. Please try again.")

if _name_ == "_main_":
    main()
```

```python
def is_prime(num):
  if num <= 1:
    return False
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False
  return True

def main():
  limit = int(input("Enter a limit: "))
  primes = [num for num in range(2, limit + 1) if is_prime(num)]
  print(f"Prime numbers up to {limit}: {primes}")

if __name__ == "__main__":
  main()
```

```python
def is_palindrome(number):
    number_str = str(number)
    return number_str == number_str[::-1]
number = 12321
if is_palindrome(number):
    print(f"{number} is a palindrome number.")
else:
    print(f"{number} is not a palindrome number.")
```

```python
def is_armstrong(number):
    num_str = str(number)
    num_digits = len(num_str)
    total = sum(int(digit) ** num_digits for digit in num_str)
    return total == number
number = 153
if is_armstrong(number):
    print(f"{number} is an Armstrong number.")
else:
    print(f"{number} is not an Armstrong number.")
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
number = 5
print(f"The factorial of {number} is {factorial(number)}")
```

```python
#Encapsulation restricts access to certain components of an object.
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    @property
    def name(self):
        return self._name

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if value > 0:
            self._age = value
        else:
            print("Age must be positive.")
person = Person("Ansh", 18)
print(person.name)
print(person.age)

person.age = 25
print(person.age)

person.age = -5
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Correct the file path if necessary
nba = pd.read_csv("/content/nbaallelo[1].csv")  # Update the path if the file is located elsewhere

# ... rest of your code ...

#Querying your dataset
current_decade=nba[nba['year_id']>2010]
print(current_decade.shape)

games_with_notes=nba[nba['notes'].notnull()]
print(games_with_notes.shape)

ers=nba[nba['fran_id'].str.endswith('ers')]
print(ers.shape)

print(nba.groupby('fran_id',sort=False)['pts'].sum())

#find Missing values
rows_without_missing_data=nba.dropna()
print(rows_without_missing_data.shape)

data_without_missing_columns=nba.dropna(axis=1)
print(data_without_missing_columns.info())

data_with_default_notes=nba.copy()
data_with_default_notes['notes'].fillna('',inplace=True)
print(data_with_default_notes['notes'].describe())

#invalid Values
print(nba[nba["pts"]==0])

#inconsistent Values
print(nba[(nba["pts"] > nba["opp_pts"]) & (nba["game_result"]!='W')].empty)
print(nba[(nba["pts"] < nba["opp_pts"]) & (nba["game_result"]!='L')].empty)

#visualizing your pandas Dataframe
print(nba[nba["fran_id"]== "Knicks"].groupby("year_id")["pts"].sum().plot())

#Combining Multiple Datasets
city_employee_count=pd.Series({"Amsterdam":5,"Tokyo":8})
city_revenues=pd.Series([4200,8000,6500],index=["Amsterdam","Toronto","Tokyo"])
city_data=pd.DataFrame({"revenue":city_revenues,"employee_count":city_employee_count})
further_city_data=pd.DataFrame({"revenue":[7000,3400],"employee_count":[2,2]},index=["New York","Barcelona"])
all_city_data=pd.concat([city_data,further_city_data],sort=False)
print(all_city_data)
```
